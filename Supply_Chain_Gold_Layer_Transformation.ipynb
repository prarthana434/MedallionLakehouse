##### **GOLD LAYER TRANSFORMATIONS AND AGGREGATIONS : SUPPLY CHAIN ANALYTICS**

```python
from pyspark.sql.functions import *
```

```python
order_lines = spark.read.parquet("Files/Supply_Chain_Silver_Layer_data/Order_lines_silver_layer/part-00000-bbf06f4c-644b-4602-98ae-06b4216d4161-c000.snappy.parquet")
# df now is a Spark DataFrame containing parquet data from "Files/Supply_Chain_Silver_Layer_data/Order_lines_silver_layer/part-00000-60c83497-18b0-4782-a09d-b4196cc75d47-c000.snappy.parquet".
display(order_lines)
```

```python
order_lines.write.mode("overwrite").parquet("Files/Supply_Chain_Gold_Layer_data/order")
display(order_lines)
```

```python
# KPI 1: Average Order value per customer, Order Qty by Order id

aov_df = order_lines.groupBy("customer_id") \
    .agg((sum("order_qty") / countDistinct("order_id")).alias("average_order_qty_per_customer"))

aov_df.write.mode("overwrite").parquet("Files/Supply_Chain_Gold_Layer_data/average_order_value")

display(aov_df)
```

```python
# KPI 2: Orders per Customer

pbc = order_lines.groupBy("customer_id") \
.agg((count("order_id")).alias("total_orders_per_customer"))

pbc.write.mode("overwrite").parquet("Files/Supply_Chain_Gold_Layer_data/orders_per_customer")

display(pbc)
```

```python
# Total items per customer

ibc = order_lines.groupBy("customer_id") \
    .agg((count("product_id")).alias("items_per_customer"))

aov_df.write.mode("overwrite").parquet("Files/Supply_Chain_Gold_Layer_data/total_items_per_customer")

display(ibc)
```
